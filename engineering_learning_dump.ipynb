# Engineering Learning Scratchbook

Short notes on development tools, language analysis, and runtime environments.

## Shell environments

A shell environment is the set of environment variables inherited when a shell process starts. Each shell process therefore holds an environment-variable snapshot.

Extensions cannot directly rewrite that snapshot in an existing shell. VS Code asks to relaunch the terminal when an extension needs to contribute new environment variables.



## Static analysis

Static-analysis tools inspect source code without executing the program. They commonly parse code into an abstract syntax tree and build a lightweight semantic model of names, scopes, types, imports, and control flow.

### Linters

A linter flags suspicious code and consistency issues, including:

- Syntax mistakes
- Unused imports or variables
- Undefined names, such as missing variables, functions, classes, or imports
- Formatting and style violations
- Error-prone patterns and overly complex functions

### Type checkers

A type checker verifies that values are used consistently with their declared or inferred types. Python annotations are optional, so tools such as Pyright and mypy perform this analysis separately from the Python runtime.

```python
def double(value: int) -> int:
    return value * 2

double("five")  # Type error: expected int
```

In statically typed languages such as C++, type checking is built into the compiler.

## Language servers

A language server provides editor features such as autocomplete, navigation, refactoring, and diagnostics. The editor communicates with it through the Language Server Protocol (LSP).

Pylance is one Python language-server implementation. Other implementations include Jedi Language Server and Python LSP Server.

```text
VS Code editor
    ↕ Language Server Protocol (LSP)
Pylance language server
    ↓
Pyright static-analysis engine
```

Pyright defines diagnostics such as `reportUndefinedVariable`. Pylance publishes those diagnostics to VS Code's Problems panel.

## Project manifests

Each language ecosystem has a conventional project manifest. Examples include `pyproject.toml` for Python, `package.json` for Node.js, `Cargo.toml` for Rust, and `go.mod` for Go.

A project manifest provides a high-level description of the project, including its identity, dependencies, runtime requirements, build and packaging configuration, executable commands, and tool settings. It usually does not list every source file.

### Python dotted import paths

Python identifies importable packages and modules with dotted paths. For a project using a `src` layout:

```text
File:        src/observatory/cli.py
Import path: observatory.cli
```

```python
import observatory.cli  # Loads src/observatory/cli.py
```
